```python
#urls stored in a list
urls = ['https://www.pluralsight.com/courses/ai-executive-briefing', 'https://www.pluralsight.com/courses/microsoft-azure-deploying-ai-solutions', 'https://www.pluralsight.com/courses/unreal-engine-creating-ai-behavior-states', 'https://www.pluralsight.com/courses/introduction-ai-navigation-systems-unreal-engine-2205', 'https://www.pluralsight.com/courses/microsoft-azure-ai-models-developing', 'https://www.pluralsight.com/courses/microsoft-azure-ai-solutions-managing', 'https://www.pluralsight.com/courses/microsoft-azure-ai-bot-framework-solution-implementing', 'https://www.pluralsight.com/courses/microsoft-azure-security-tools-protect-ai-solutions', 'https://www.pluralsight.com/courses/microsoft-azure-operationalizing-ai-solutions', 'https://www.pluralsight.com/courses/microsoft-azure-developer-creating-integrating-ai-services-update', 'https://www.pluralsight.com/courses/microsoft-azure-identifying-security-requirements-ai-solution', 'https://www.pluralsight.com/courses/smart-analytics-machine-learning-ai-gcp', 'https://www.pluralsight.com/courses/unity-building-ai-driven-characters-using-playmaker', 'https://www.pluralsight.com/courses/tq-ai-channel-introduction', 'https://www.pluralsight.com/courses/big-data-ldn-session-80', 'https://www.pluralsight.com/courses/microsoft-azure-optimizing-ai-solutions', 'https://www.pluralsight.com/courses/ai-algorithms-kismet-udk-1248', 'https://www.pluralsight.com/courses/devintersection-azureai-session-33', 'https://www.pluralsight.com/courses/devintersection-azureai-session-46', 'https://www.pluralsight.com/courses/microsoft-ignite-session-14', 'https://www.pluralsight.com/courses/devintersection-azureai-session-32', 'https://www.pluralsight.com/courses/swords-shovels-character-controller-ai', 'https://www.pluralsight.com/courses/microsoft-ignite-session-47', 'https://www.pluralsight.com/courses/big-data-ldn-session-04', 'https://www.pluralsight.com/courses/big-data-ldn-session-161', 'https://www.pluralsight.com/courses/microsoft-ignite-session-13', 'https://www.pluralsight.com/courses/codemash-session-111', 'https://www.pluralsight.com/courses/big-data-ldn-session-87', 'https://www.pluralsight.com/courses/devintersection-azureai-session-16', 'https://www.pluralsight.com/courses/microsoft-ignite-session-108', 'https://www.pluralsight.com/courses/devintersection-azureai-session-29', 'https://www.pluralsight.com/courses/microsoft-ignite-session-73', 'https://www.pluralsight.com/courses/tq-artificial-intelligence', 'https://www.pluralsight.com/courses/big-data-ldn-session-71', 'https://www.pluralsight.com/courses/big-data-ldn-session-65', 'https://www.pluralsight.com/courses/microsoft-ignite-session-109', 'https://www.pluralsight.com/courses/devintersection-azureai-session-02', 'https://www.pluralsight.com/courses/devintersection-azureai-session-01', 'https://www.pluralsight.com/courses/big-data-ldn-session-155', 'https://www.pluralsight.com/courses/microsoft-ignite-session-110', 'https://www.pluralsight.com/courses/big-data-ldn-session-54', 'https://www.pluralsight.com/courses/big-data-ldn-session-133', 'https://www.pluralsight.com/courses/big-data-ldn-session-48', 'https://www.pluralsight.com/courses/big-data-ldn-session-60', 'https://www.pluralsight.com/courses/microsoft-azure-choosing-appropriate-services-features', 'https://www.pluralsight.com/courses/microsoft-azure-intelligent-edge-designing', 'https://www.pluralsight.com/courses/big-data-ldn-session-24', 'https://www.pluralsight.com/courses/ng-conf-19-future-machine-learning-javascript', 'https://www.pluralsight.com/courses/devintersection-azureai-session-31', 'https://www.pluralsight.com/courses/devintersection-azureai-session-14', 'https://www.pluralsight.com/courses/devintersection-azureai-session-28', 'https://www.pluralsight.com/courses/devintersection-azureai-session-19', 'https://www.pluralsight.com/courses/devintersection-azureai-session-45', 'https://www.pluralsight.com/courses/artificial-intelligence-big-picture', 'https://www.pluralsight.com/courses/devintersection-azureai-session-30', 'https://www.pluralsight.com/courses/devintersection-azureai-session-13', 'https://www.pluralsight.com/courses/devintersection-azureai-session-43', 'https://www.pluralsight.com/courses/devintersection-azureai-session-17', 'https://www.pluralsight.com/courses/devintersection-azureai-session-18', 'https://www.pluralsight.com/courses/devintersection-azureai-session-15', 'https://www.pluralsight.com/courses/devintersection-azureai-session-34', 'https://www.pluralsight.com/courses/unity-navigation-fundamentals', 'https://www.pluralsight.com/courses/arnold-maya-rendering-volumes-effects', 'https://www.pluralsight.com/courses/microsoft-azure-developing-ml-pipelines', 'https://www.pluralsight.com/courses/microsoft-cognitive-services-text-analytics-api', 'https://www.pluralsight.com/courses/microsoft-ignite-session-30', 'https://www.pluralsight.com/courses/devintersection-azureai-session-12', 'https://www.pluralsight.com/courses/big-data-ldn-session-74', 'https://www.pluralsight.com/courses/microsoft-azure-search-solution-implementing', 'https://www.pluralsight.com/courses/devintersection-azureai-session-22', 'https://www.pluralsight.com/courses/microsoft-azure-cognitive-services-text-speech-api', 'https://www.pluralsight.com/courses/that-conference-2019-session-05', 'https://www.pluralsight.com/courses/unity-2d-fundamentals-obstacles-enemies', 'https://www.pluralsight.com/courses/deploying-machine-learning-solutions', 'https://www.pluralsight.com/courses/production-machine-learning-systems', 'https://www.pluralsight.com/courses/controlling-an-agents-actions-massive-prime-1627', 'https://www.pluralsight.com/courses/how-google-does-machine-learning', 'https://www.pluralsight.com/courses/microsoft-azure-textual-content-search-enabling', 'https://www.pluralsight.com/courses/microsoft-cognitive-services-machine-learning', 'https://www.pluralsight.com/courses/big-data-ldn-session-145', 'https://www.pluralsight.com/courses/microsoft-azure-cognitive-services-text-translator-api', 'https://www.pluralsight.com/courses/big-data-ldn-session-128', 'https://www.pluralsight.com/courses/microsoft-cognitive-services-bing-custom-search-api', 'https://www.pluralsight.com/courses/devintersection-azureai-session-10', 'https://www.pluralsight.com/courses/devintersection-azureai-session-07', 'https://www.pluralsight.com/courses/microsoft-ignite-session-42', 'https://www.pluralsight.com/courses/big-data-ldn-session-93', 'https://www.pluralsight.com/courses/big-data-ldn-session-05', 'https://www.pluralsight.com/courses/end-end-machine-learning-tensorflow-gcp', 'https://www.pluralsight.com/courses/big-data-ldn-session-31', 'https://www.pluralsight.com/courses/scratch-2-0-making-games', 'https://www.pluralsight.com/courses/genetic-algorithms-genetic-programming', 'https://www.pluralsight.com/courses/sharepoint-conference-2019-session-44', 'https://www.pluralsight.com/courses/big-data-ldn-session-106', 'https://www.pluralsight.com/courses/ng-conf-19-machine-learning-tensorflow-js', 'https://www.pluralsight.com/courses/microsoft-ignite-session-101', 'https://www.pluralsight.com/courses/big-data-aws-big-picture', 'https://www.pluralsight.com/courses/microsoft-azure-cognitive-services-recognition-bing-speech-api', 'https://www.pluralsight.com/courses/google-cloud-platform-designing-scalable-data-architectures', 'https://www.pluralsight.com/courses/aws-comprehend-analyzing-text', 'https://www.pluralsight.com/courses/microsoft-ignite-session-46', 'https://www.pluralsight.com/courses/securing-data-aws', 'https://www.pluralsight.com/courses/aws-comprehend-analyzing-text', 'https://www.pluralsight.com/courses/build-train-deploy-first-neural-network-tensorflow', 'https://www.pluralsight.com/courses/turn-based-game-unity-playmaker-2160', 'https://www.pluralsight.com/courses/game-maker-studio-2-making-your-first-game', 'https://www.pluralsight.com/courses/pytorch-building-deep-learning-models', 'https://www.pluralsight.com/courses/serverless-analytics-aws', 'https://www.pluralsight.com/courses/introduction-tensorflow', 'https://www.pluralsight.com/courses/understanding-ethical-legal-issues-data-science', 'https://www.pluralsight.com/courses/sccm-2007-pt2', 'https://www.pluralsight.com/courses/machine-learning-business-professionals', 'https://www.pluralsight.com/courses/google-cloud-automl-designing-implementing-solutions', 'https://www.pluralsight.com/courses/sql-bigquery-ml-building-machine-learning-models', 'https://www.pluralsight.com/courses/building-features-image-data', 'https://www.pluralsight.com/courses/creating-voice-chatbots-work-everywhere', 'https://www.pluralsight.com/courses/unity-blender-fbx-workflows', 'https://www.pluralsight.com/courses/sequence-models-time-series-natural-language-processing', 'https://www.pluralsight.com/courses/microsoft-azure-reducing-complexity-data', 'https://www.pluralsight.com/courses/google-dialogflow-chatbots', 'https://www.pluralsight.com/courses/microsoft-azure-stack-solutions-architectural-patterns', 'https://www.pluralsight.com/courses/unreal-engine-4-fundamentals', 'https://www.pluralsight.com/courses/microsoft-azure-cognitive-services-custom-vision-api', 'https://www.pluralsight.com/courses/unity-rigging-2d-character-ik', 'https://www.pluralsight.com/courses/google-machine-learning-apis-designing-implementing-solutions', 'https://www.pluralsight.com/courses/how-machine-learning-works', 'https://www.pluralsight.com/courses/microsoft-azure-cognitive-services-custom-vision-api', 'https://www.pluralsight.com/courses/unity-rigging-2d-character-ik', 'https://www.pluralsight.com/courses/google-machine-learning-apis-designing-implementing-solutions', 'https://www.pluralsight.com/courses/building-features-text-data', 'https://www.pluralsight.com/courses/animate-cc-lip-syncing-2017', 'https://www.pluralsight.com/courses/microsoft-bot-framework-building-bots-getting-started-2019', 'https://www.pluralsight.com/courses/google-cloud-functions-fundamentals', 'https://www.pluralsight.com/courses/unity-mobile-game-dev-enemy-ai-waypoints-641', 'https://www.pluralsight.com/courses/unity-mobile-game-dev-final-assembly-636', 'https://www.pluralsight.com/courses/animate-cc-lip-syncing', 'https://www.pluralsight.com/courses/2d-racing-games-unity-volume-4-1311', 'https://www.pluralsight.com/courses/ps-live-2018']


#empty lists for scrapped data
titles = []
authors = []
levels = []
updates = []
durations = []
descriptions = []
lessons = []
ratings = []

#start time
start_time = time()
requests = 0

for url in urls:
    response = get(url)
    
    sleep(randint(8,15))
    
    requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    
    if response.status_code != 200:
        print('status code not 200')

    if requests > 139:
        print('too many requests!')
        break
    
    
    html_soup = BeautifulSoup(response.text, 'html.parser')
    
    title = html_soup.find('div', class_='title section')
    titles.append(title.h1.text.strip())
    
    author = html_soup.find('h5', class_='title--alternate')
    authors.append(author.a.text)
    
    level = html_soup.find('div', class_='course-info__row--right difficulty-level')
    levels.append(level.get_text().strip())
    
    updated = html_soup.find('div', string='Updated', class_='course-info__row--left')
    updates.append(updated.find_next_sibling("div").get_text().strip())
    
    duration = html_soup.find('div', string='Duration', class_='course-info__row--left')
    durations.append(duration.find_next_sibling("div").get_text().strip())
    
    description = html_soup.find('div', class_='text-component')
    descriptions.append(description.get_text())
    
    lesson = html_soup.find_all('span', class_='accordion-content__row__title')
    lessons.append(len(lesson))
    
    #place for rating calc
    rating = html_soup.find('div', class_='course-info__row--right course-info__row--rating')
    if rating:
        rating_full = len(rating.find_all('i', class_='fa fa-star'))
        rating_half = len(rating.find_all('i', class_='fa fa-star-half-o'))
        rating_blank = len(rating.find_all('i', class_='fa fa-star gray'))

        rating_total = rating_full*1 + rating_half*0.5 + rating_blank*0
        ratings.append(rating_total)
        
    else:
        ratings.append('none')
    
    
    print('New item added: ', title.h1.text.strip())

   
```

```python
import pandas as pd

data = pd.DataFrame({'Course': titles,'Author': authors,'Level':levels,'Rating':ratings,'Duration':durations,'Lessons':lessons,
                     'Updated':updates,
                     'Description':descriptions})

data.head()
```
